# Simulación de Fluidos con WaterLily.jl
Ejecuta la celda de abajo para ver la generación del GIF de vorticidad.

```julia
using WaterLily, Plots, Logging
disable_logging(Logging.Debug)

cID = "2DCircle"

function circle(n, m; Re=250, U=1, mem=Array)
    radius, center = m/8, m/2
    body = AutoBody((x, t) -> √sum(abs2, x .- center) - radius)
    Simulation((n, m), (U, 0), radius; ν=U*radius/Re, body, mem)
end

# Inicializar
sim = circle(3*2^6, 2^7)
WaterLily.logger(cID)

# Generar GIF (Esto tardará unos minutos la primera vez)
sim_gif!(sim, duration=10, clims=(-5, 5), plotbody=true)
```
